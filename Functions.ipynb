## 1. What is the difference between a function and a method in Python?

  - **Function**: A function is aresuable block of code defined using the **"def"** keyword and can be called anywhere in the program.

  - **Method**: A method is similar to a function but is associated with an object and can only be called using the object. Methods are defined inside **classes** and **operate** on the object's data.

```python
# Function
def greet(name):
    return f"Hello, {name}!"
print(greet("Ravi"))  # Calling a function



# Method
class colour:
    def __init__(self, colour):
        self.colour = colour

    def greet(self):
        return f"Hello, {self.colour}!"

c = colour("Green")
print(c.greet())  # Calling a method using an object
```

## 6. Explain the concept of generators in Python and how they are defined.

  - **Generators** in python are a special type of iterator that yield values one at a time using yield instead of return. It defined using "Functions with **yield**".

```python
def count_up(n):
    for i in range(n):
        yield i

gen = count_up(5)
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```

## 7. What are the advantages of using generators over regular functions?

  Advantage of using Generators over regular function are as follow:-
  - **Memory Efficient**: Generators do not store all values in memory.
  - **Lazy Evaluation**: Generates values on demand, improving performance for large data.
  - **Simplifies Code**: No need to manually manage iterators.


```python
def infinite_counter():
    n = 0
    while True:
        yield n
        n += 1

counter = infinite_counter()
print(next(counter))
print(next(counter))
print(next(counter))
```

## 10. What is the difference between map(), reduce(), and filter() functions in Python?

  - **map()**: Applies a function to each item in an iterable.
  - **filter()**: Filters elements based on a condition.
  - **reduce()**: Performs a cumulative computation on an iterable.


```python
from functools import reduce

numbers = [1, 2, 3, 4, 5]

# map: Squaring each element
squares = list(map(lambda x: x ** 2, numbers))
print(squares)  # Output: [1, 4, 9, 16, 25]

# filter: Filtering even numbers
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)  # Output: [2, 4]

# reduce: Summing all elements
sum_all = reduce(lambda x, y: x + y, numbers)
print(sum_all)  # Output: 15
```

## 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.


```python
def sum_even_numbers(lst):
    return __builtins__.sum(num for num in lst if num % 2 == 0)

# Example
numbers = [1, 2, 3, 4, 5, 6]
print(sum_even_numbers(numbers))
```

## 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

# Example: Check all primes from 1 to 200
primes = [n for n in range(1, 201) if is_prime(n)]
print(primes)
```

## 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration
        self.count += 1
        self.a, self.b = self.b, self.a + self.b
        return self.a

# Example
fib = FibonacciIterator(10)
print(list(fib))  # Output: First 10 Fibonacci numbers
```

## 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example
for power in powers_of_2(5):
    print(power)  # Output: 1, 2, 4, 8, 16, 32
```

## 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def remove_vowels(s):
    return "".join(filter(lambda c: c.lower() not in "aeiou", s))

# Example
text = "Ravi Ranjan Kumar"
print(remove_vowels(text))
```
